```python
#pra variar, vou dividir em 9 grupos:
Q1 = df_pib['PIB2020'].quantile(0.1)
Q2 = df_pib['PIB2020'].quantile(0.2)
Q3 = df_pib['PIB2020'].quantile(0.3)
Q4 = df_pib['PIB2020'].quantile(0.4)
Q5 = df_pib['PIB2020'].quantile(0.5)  # Mediana (Quintil 5)
Q6 = df_pib['PIB2020'].quantile(0.6)
Q7 = df_pib['PIB2020'].quantile(0.7)
Q8 = df_pib['PIB2020'].quantile(0.8)

print('Primeiro nono: ', Q1)
print('Segundo nono: ', Q2)
print('Terceiro nono: ', Q3)
print('Quarto nono: ', Q4)
print('Mediana (Quintil 5): ', Q5)
print('Sexto nono: ', Q6)
print('Sétimo nono: ', Q7)
print('Oitavo nono: ', Q8)
```

```python
#Agora vai ser mais fácil dar nome ao pessoal
df_pib['QuintilRank'] = pd.qcut(
    df_pib['PIB2020'],
    q=9,
    labels=[
        "paupérrimo",
        "muito pobre",
        "pobre",
        "médio pobre",
        "médio",
        "médio rico",
        "rico",
        "muito rico",
        "riquíssimo",
        
    ])
```

```python
df_pib
```

```python
df_pib.tail()
```

```python
# verificando quantos municipios ficaram em cada grupo
contagem_quintis = df_pib['QuintilRank'].value_counts()
```

```python
contagem_quintis
```

```python
# criar a média de cada quintil com base no valor do PIB2020
media_quintil = df_pib.groupby('QuintilRank')['PIB2020'].mean()
```

```python
media_quintil
```

```python
#Agora vou cirar um gráfico de barras com essas médias
media_quintil = df_pib.groupby('QuintilRank')['PIB2020'].mean()

# Plotar o gráfico de barras das médias por quintil
media_quintil.plot(kind='bar', figsize=(10, 6))
plt.title('Média do PIB2020 por Quintil')
plt.xlabel('Quintil')
plt.ylabel('Média do PIB de 2020')
plt.xticks(rotation=45)  # Rotacionar os rótulos do eixo x, se necessário
plt.show()
```

```python
#finalmente um histograma, para verificarmos a frequencia que os valores aparecem em cada quintil 
dados = df_pib['PIB2020']

# Suponha que 'media_quintil' já contenha as médias dos quintis calculadas anteriormente.
# Transforme a série em uma lista de valores.
medias_quintis = media_quintil.values

# Plotar o histograma com as médias dos quintis como bins
plt.figure(figsize=(10, 6))  # Tamanho da figura
plt.hist(dados, bins=medias_quintis, color='skyblue', edgecolor='black')
plt.title('Histograma do PIB ')
plt.xlabel('PIB em 2020')
plt.ylabel('Frequência')
plt.show()
```
