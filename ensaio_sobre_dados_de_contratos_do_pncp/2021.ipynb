```python
    #a2021 = aqui tenho núm do contrato e do forncedor
##pode se colocar todas a colunas que quer trabalhar
    #groupby = aqui agrupou por fornecedor
    #sum = aqui faz a contagem de contratos únicos por fornecedor
    #sort_values = aqui ordena pelo núm de contrato existente
    #faz a soma pelo valor valorGlobal
    #head(10) = os 10 maiores contratos em valores
    #style.format('{:.2f}') = este comando formata o dataframe inteiro para que todos os números tenha
##no máximo 2 casas depois da vírgula (se 3f, tres casas; se 4f, casas etc.)

a2021[['numeroControlePNCP', 'niFornecedor', 'valorGlobal']]\
    .groupby(by= ['niFornecedor'])\
    .sum('valorGlobal')\
    .sort_values(by= ['valorGlobal'], ascending= False)\
    .head(10)\
    .style.format('{:.2f}')
```
