```python
evadidos = dados_evasao[dados_evasao['SITUACAO'].str.lower() == 'evadido']
```

```python
evadidos_por_ano = evadidos.groupby('ANO_INGRESSO').size()
```

```python
evadidos_por_ano
```

```python
plt.plot(evadidos_por_ano.index, evadidos_por_ano.values, marker='o', linestyle='-')
plt.xlabel('Ano')
plt.ylabel('Número de Evadidos')
plt.title('Linha do Tempo de Evadidos por Ano de ingresso')
plt.grid(True)
plt.show()
```

```python
#Gráfico de barras
evadidos_por_ano.plot(kind='bar', legend=False)

plt.xlabel('Ano de Ingresso')
plt.ylabel('Número de Evadidos')
plt.title('Número de Evadidos por Ano de Ingresso')
plt.grid(axis='y')  # Apenas grade no eixo Y
plt.show()
```

```python
dados_evasao['ANO_INGRESSO']
```

```python
contagem_por_ano = dados_evasao.groupby('ANO_INGRESSO').size().reset_index(name='Contagem')
contagem_por_ano = contagem_por_ano.sort_values(by='ANO_INGRESSO')
print(contagem_por_ano)
```

```python
contagem_total = contagem_por_ano['Contagem'].sum()
total_por_ano = pd.DataFrame({'ANO_INGRESSO': ['Total'], 'Contagem': [contagem_total]})

contagem_com_total = pd.concat([contagem_por_ano, total_por_ano])
contagem_com_total = contagem_com_total[contagem_com_total['ANO_INGRESSO'] != 'Total']
contagem_com_total.plot(x='ANO_INGRESSO', y='Contagem', kind='bar', legend=False)

plt.xlabel('Ano de Ingresso')
plt.ylabel('Número de Linhas')
plt.title('Contagem por Ano de Ingresso')
plt.show()
```

```python
# Gráfico de barra - falta incluir informação de proporção
ax = contagem_com_total.plot(x='ANO_INGRESSO', kind='bar', legend=True, figsize=(10, 6))
evadidos_por_ano.plot(x='ANO_INGRESSO', y='Contagem', kind='bar', color='red', legend=True, ax=ax)

plt.xlabel('Ano de Ingresso')
plt.ylabel('Total')
plt.title('Total de ingressos e Evadidos por Ano de Ingresso')
plt.show()
```
