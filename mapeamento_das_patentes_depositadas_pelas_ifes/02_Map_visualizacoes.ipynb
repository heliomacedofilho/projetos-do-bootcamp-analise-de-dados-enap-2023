# Tratamento dos dados

## Carregar o Dataframe

```python
# Carregar o arquivo CSV
data = pd.read_csv('data_patentes.csv')
data.head(10)
```

## Datas

```python
#Transformar colunas data_publicacao, data_deposito e data_prioridade em datas
data['data_publicacao'] = pd.to_datetime(data['data_publicacao'], format='%d/%m/%Y')
data['data_deposito'] = pd.to_datetime(data['data_deposito'], format='%d/%m/%Y')
data['data_prioridade'] = pd.to_datetime(data['data_prioridade'], format='%d/%m/%Y')
```
